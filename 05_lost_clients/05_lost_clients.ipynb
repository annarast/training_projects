### Обработка пропусков после объединения

После объединения таблиц появились пропуски в столбцах из таблиц internet и phone. Это связано с тем, что не все клиенты пользуются услугами интернета и телефонной связи. Для заполнения пропусков используем заглушки, которые позволят понимать, пользуется клиент услугой или нет.

```python
for column in internet.columns[2:]:
    data[column] = data[column].fillna('No')
data['internet_service'] = data['internet_service'].fillna('no_internet')    
data['multiple_lines'] = data['multiple_lines'].fillna('no_phone')
```

```python
data.info()
```

### Создание новых колонок

Для анализа данных и определения признаков, потенциально влияющих на решение клиентов уйти, создадим столбец "target", в котором как 1 будут обозначены ушедшие клиенты, а как 0 - действующие.

```python
data['target'] = np.where(data['end_date'] =="1900-01-01", 0, 1)
```

Для целей анализа и обучения моделей интересны не даты заключения контрактов, а продолжительность сотрудничества клиента с компанией. Добавим в датасет столбцы с расчётом года начала пользования услугами компании и продолжительностью пользования услугами.

```python
data['len_month'] = np.where(data['end_date'] == '1900-01-01', 
                ((pd.to_datetime('2020-02-01', format='%Y-%m-%d') - data['begin_date'])/np.timedelta64(1,'M')).astype(int), 
                ((data['end_date'] - data['begin_date'])/np.timedelta64(1,'M')).astype(int))
```

```python
data['len_year'] = np.where(data['end_date'] == '1900-01-01', 
                       2020 - data['begin_date'].dt.year, 
                       data['end_date'].dt.year - data['begin_date'].dt.year)
```

```python
data['begin_year'] = data['begin_date'].dt.year
```

Также важной для анализа поведения клиента может быть информация о том, пользуется он только интернетом, только телефонной связью или и тем, и другим. Поэтому добавим столбец с этими данными.  
Помним, что если при объединении таблиц появились пропуски в столбцах, связанных с интернетом или в столбцах, связанных с телефонной связью - это значит, что клиент этими видами связи не пользуется.  
Введём следующие категории:  
"both" - клиент пользуется и телефонной связью, и интернетом,  
"phone" - клиент пользуется только интернетом,  
"internet" - клиент пользуется толькр телефонной связью,  
"no_one" - клиент не пользуется ни тем, ни другим (если вдруг такие есть)

```python
data['services'] = np.where((data['internet_service'] != 'no_internet')&(data['multiple_lines'] != 'no_phone'), 'both',
                            np.where((data['internet_service'] == 'no_internet')&(data['multiple_lines'] == 'no_phone'), 'no_one',
                                     np.where(data['internet_service'] == 'no_internet', 'phone',
                                              np.where(data['multiple_lines'] == 'no_phone', 'internet', 'error'))))
```

```python
data['services'].value_counts(normalize=True)
```

Получается, что 68% клиентов пользуются и связью, и интернетом, 21 % - только связью, 10% - только интернетом.

### Логистическая регрессия

```python
#подготавливаем признаки для логистической регрессии, удаляя те, у которых высокая корреляция с другими (0,7 и больше)
features_train_lr = features_train.drop(['internet_service_Fiber optic', 
                                 'internet_service_no_internet', 
                                 'streaming_tv_Yes', 
                                 'streaming_movies_Yes',
                                 'multiple_lines_Yes'], axis=1)
```

```python
%%time
pipe_lr = Pipeline([('scl', StandardScaler()),
                ('clf', LogisticRegression(random_state = 120922))
                ])
params_lr  = {'clf__penalty':['l1','l2','elasticnet'],
                'clf__C':range(1,50)
             }

grid_lr = GridSearchCV(estimator = pipe_lr,
                       param_grid=params_lr,
                       scoring ='roc_auc',
                       cv=5,
                       n_jobs=4
                      )
grid_lr.fit(features_train_lr, target_train)
print('Лучшие гиперпараметры логистической регрессии', grid_lr.best_params_)
print('Лучшая ROC AUC логистической регрессии', '{:.5f}'.format(grid_lr.best_score_))


best_params_list.append(grid_lr.best_params_)
roc_auc_list.append(grid_lr.best_score_)

print('Accuracy:', '{:.5f}'.format(accuracy_score(grid_lr.best_estimator_.named_steps['clf'].predict(features_train_lr), 
                                                  target_train)))
accuracy_list.append(accuracy_score(grid_lr.best_estimator_.named_steps['clf'].predict(features_train_lr), 
                                                  target_train))

#построение графика ROC кривой 
probs_lr = grid_lr.best_estimator_.named_steps['clf'].predict_proba(features_train_lr)
positive_probs_lr = probs_lr[:, 1]
fpr_lr, tpr_lr, treshold_lr = roc_curve(target_train, positive_probs_lr)

plt.plot(fpr_lr, tpr_lr, color='blue',
         label='ROC кривая Логистическая регрессия')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая')
plt.legend(loc="lower right")
```

```python
#определение важности признаков
feature_importance=pd.DataFrame({'feature':list(features_train_lr.columns),
                                 'feature_importance':grid_lr.best_estimator_.named_steps['clf'].coef_[0]})
feature_importance.sort_values('feature_importance',ascending=False).style.bar()
```

Для логистической регрессии почти все признаки довольно важны.
Обучим модель, пошагово убирая признаки с небольшой важностью, чтобы определить, улучшится ли метрика.

```python
%%time
#изменение roc_auc при исключении части незначимых признаков
steps = [0.05, 0.1, 0.2, 0.3]
steps_list = []
features_list = []
for step in steps:
    unimportant_features_lr = feature_importance[(feature_importance['feature_importance']<step)&
                                                (feature_importance['feature_importance']>-step)]['feature'].tolist()
    pipe_lr = Pipeline([('scl', StandardScaler()),
                ('clf', LogisticRegression(random_state = 120922))
                ])
    params_lr  = {'clf__penalty':[grid_lr.best_params_['clf__penalty']],
                'clf__C':[grid_lr.best_params_['clf__C']]
             }

    grid_lr = GridSearchCV(estimator = pipe_lr,
                       param_grid=params_lr,
                       scoring ='roc_auc',
                       cv=5,
                       n_jobs=4
                      )
    grid_lr.fit(features_train.drop(unimportant_features_lr, axis=1), target_train)

    steps_list.append(grid_lr.best_score_)


steps_scores = pd.DataFrame({'steps':steps,
                             'roc-auc':steps_list})
#выделяем зелёным цветом наибольшее значения метрики
display(steps_scores.style.highlight_max(color='yellowgreen', subset='roc-auc'))
```
